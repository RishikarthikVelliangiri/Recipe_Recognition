```python
import pandas as pd

# your existing DataFrame
# df = pd.read_csv(...)

# 1️⃣ Define each bucket as a Python list
soups = [
    'Creamy White Chili','Cheeseburger Soup','Cauliflower Soup',
    'Rustic Italian Tortellini Soup','Stuffed Pepper Soup','Cheesy Ham Chowder',
    'Best Ever Potato Soup','Chicken Wild Rice Soup','Comforting Chicken Noodle Soup',
    'Contest-Winning New England Clam Chowder','Quick Cream of Mushroom Soup',
    'White Bean Chicken Chili','Simple Taco Soup','Black Bean ‘n’ Pumpkin Chili',
    'Cheeseburger Paradise Soup'
]

breads = [
    'Best Ever Banana Bread','Basic Homemade Bread','Buttery Cornbread',
    'Gluten-Free Banana Bread','Lemon Blueberry Bread','Special Banana Nut Bread',
    'Pumpkin Bread','Basic Banana Muffins'
]

cakes = [
    'Pumpkin Spice Cupcakes with Cream Cheese Frosting','Zucchini Cupcake',
    'Moist Chocolate Cake','Sandy’s Chocolate Cake','Pineapple Orange Cake',
    'Hot Milk Cake','Chunky Apple Cake','Chocolate Guinness Cake',
    'Chocolate-Strawberry Celebration Cake','Mamaw Emily’s Strawberry Cake'
]

bars = [
    'Frosted Banana Bars','Rhubarb Custard Bars','Cherry Bars',
    'Big Soft Ginger Cookies','Chocolate Chip Oatmeal Cookies',
    'First-Place Coconut Macaroons','Banana Bars with Cream Cheese Frosting',
    'Pumpkin Bars'
]

pies = [
    'Apple Pie','Favorite Dutch Apple Pie','Fluffy Key Lime Pie',
    'Caramel-Pecan Cheesecake Pie'
]

casseroles = [
    'Zucchini Pizza Casserole','Baked Spaghetti',
    'Chicken Penne Casserole','Twice-Baked Potato Casserole'
]

pasta = [
    'Traditional Lasagna','Seafood Lasagna','Taco Lasagna',
    'Slow-Cooker Lasagna','Ravioli Lasagna','Vegetarian Linguine'
]

salads_sides = [
    'Macaroni Coleslaw','Creamy Grape Salad','Corn Pudding',
    'Simple Au Gratin Potatoes','Creamy Coleslaw','Bacon Macaroni Salad',
    'Strawberry Pretzel Salad'
]

breakfast = [
    'Amish Breakfast Casserole','Fluffy Pancakes',
    'Brown Sugar Oatmeal Pancakes','Blueberry French Toast'
]

dessert_specialties = [
    'Winning Apple Crisp','Peanut Butter Chocolate Dessert',
    'Pineapple Pudding Cake','Peanut Butter Cup Cheesecake',
    'Caramel Heavenlies','Chocolate Caramel Candy',
    'Homemade Peanut Butter Cups','Easy Peanut Butter Fudge',
    'Tennessee Peach Pudding'
]

main_dishes = [
    'Favorite Chicken Potpie','Flavorful Chicken Fajitas','Enchilada Casser-Ole!',
    'Li’l Cheddar Meat Loaves','Mom’s Meat Loaf','Baked Mushroom Chicken',
    'Forgotten Jambalaya','Flavorful Pot Roast','Shrimp Scampi',
    'Garlic Beef Enchiladas','Bruschetta Chicken','Grilled Huli Huli Chicken',
    'Asian Chicken Thighs','Pork Chops with Scalloped Potatoes',
    'Chicken and Dumplings','Smothered Chicken Breasts','Baked Tilapia',
    'Easy Chicken Enchiladas','Lime Chicken Tacos','Skillet Shepherd’s Pie',
    'Porcupine Meatballs','Ham and Swiss Sliders','Teriyaki Chicken Thighs',
    'Egg Roll Noodle Bowl','Creamy Macaroni and Cheese'
]

# 2️⃣ Build a single mapping dict
category_map = {}
category_map.update({r: 'Soups & Chilis'       for r in soups})
category_map.update({r: 'Breads & Muffins'     for r in breads})
category_map.update({r: 'Cakes & Cupcakes'     for r in cakes})
category_map.update({r: 'Bars & Cookies'       for r in bars})
category_map.update({r: 'Pies & Tarts'         for r in pies})
category_map.update({r: 'Casseroles & Bakes'   for r in casseroles})
category_map.update({r: 'Pasta & Lasagna'      for r in pasta})
category_map.update({r: 'Salads & Sides'       for r in salads_sides})
category_map.update({r: 'Breakfast'            for r in breakfast})
category_map.update({r: 'Dessert Specialties'  for r in dessert_specialties})
category_map.update({r: 'Main Dishes'          for r in main_dishes})

# 3️⃣ Apply it to your DataFrame
df['category'] = df['recipe_name'].map(category_map)

# Now every one of your 100 recipes will have an explicit category.
print(df[['recipe_name','category']].head(10))
```
